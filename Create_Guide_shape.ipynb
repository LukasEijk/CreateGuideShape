```python
from guidelineLib import *
```

# Set mathematical functions

```python
x = smp.symbols('x')
```

```python

# y = .5 * x ** 2  # 0.8* x**2
z = 0.0 * x + 1

# one can use Piecewise functions as well
# y = smp.Piecewise(
#    (-(x+1), x < -1),
#    ((x+1)**2  , x >= -1)
# )

y = 0.001 * (x - 5) * (x - 4) ** 2 * (x - 3) * (x - 2) * (x - 1) * x * (x + 1) * (x + 2) * (x + 3)

width = .5 + 0 * x  # smp.Abs(1.*smp.cos(x)) +.5 #(smp.sin(w) +1.5) *3
height = 0.2*x + 1

```

## set Parameters

```python
start = -2
stop = 5

max_step: float = 3.
max_xStep = max_step
delta = .05

funcs = [y, z, width, height]
dfuncs = [smp.diff(func, x) for func in funcs]
```

# Create Guideline

```python
x_list = create_xList_ProgressBar(funcs, dfuncs, x, start, stop, max_xStep, delta)

rects = create_coordlist_for_off(x_list, x, funcs)

write_blender_off(rects)
```
